# Creating Dummy variables for qualitative variables

```python
df=pd.get_dummies(df)
```

```python
df.head()
```

# While creating a dummy variable we take (no.of different variables-1) no. of variables as the other variables give us all the required info about the deleted variable.

# Eg: 
# Here in the waterbody there are four different variables. Hence we create dummies for 3 variables. 

# Here if there's a value 1 in a variable then that particular variable exists. 

# If all the 3 dummies created have the value 0, then the deleted variable exists.

```python
del df['airport_NO']
del df['waterbody_None']
```

```python
df.head()
```

```python
df.corr()
```

##### For the above correlation matrix observe the correlation between the independent variables. If two independent variables have '>0.8' or '<-0.8' we delete one of those independent variables to avoid multicollinearity.
##### To choose between one of the two highly correlated independent variables we take the correlation of each of those indpendent variables to that of the dependent variable. The variable which has lower correlation with the dependent variable is deleted.
##### Eg:
##### In the above correlation matrix the independent variables 'Parks' & 'air_quality' have high correlation (0.915544)
##### We compare each of the variables' correlation with the dependent variable.
##### The correlation of Parks with price is '-0.391574'
##### The correlation of air_quality with price is '-0.429300'
##### The air_quality variable has higher negative correlation than Parks. Hence, we delete 'Parks'

```python
del df['parks']
df.head()
```

# Ridge Regression

## For Ridge Regression we always have to standardize the variables 

## We create scaler object to store the scaling of 'x' variables

```python
from sklearn import preprocessing
```

```python
scaler= preprocessing.StandardScaler().fit(x_train)
```

```python
x_train_s = scaler.transform(x_train)
```

```python
x_test_s= scaler.transform(x_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r= Ridge(alpha = 0.5)
lm_r.fit(x_train_s, y_train)
```

```python
r2_score(y_test, lm_r.predict(x_test_s))
```

```python
from sklearn.model_selection import validation_curve
```
